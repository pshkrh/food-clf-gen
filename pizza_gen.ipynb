```python
from diffusers import DDPMPipeline
from PIL import Image
import torch
from IPython.display import display
```

```python
pipeline = DDPMPipeline.from_pretrained('./ddpm-pizza-256-v3', use_safetensors=True).to('cuda')
images = pipeline(batch_size=16, generator=torch.manual_seed(0)).images

for idx, img in enumerate(images, start=1):
    display(img)
    img.save(f'pizza-generated-{idx}.jpg')
```
